```python
import sys
sys.path.insert(1, "../src/")

import numpy as np
from ca import CA, CA_STAR
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
```

```python
# Initialise
n = 100
enc = '011010'
ca = CA_STAR(n = n,
             enc = enc,
             r = 2,
             density = 0.8)

ca.env
```

```python
ca.run(steps=200)
```

```python
# ca.env_history
```

```python
# Plot

im = np.zeros((len(ca.env_history), ca.n))
for i in range(len(ca.env_history)):
    im[i,:] = ca.env_history[i]

fig = plt.gcf()
plt.imshow(im, cmap='gray')
plt.title(enc)
plt.axis('off')
plt.savefig('../imgs/test_011010/'+str(2)+'.png')
plt.show()
```
